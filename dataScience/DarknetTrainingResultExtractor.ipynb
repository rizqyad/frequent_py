```python
# open file
modelname = "Y4DB5xx"
try:
    file = open("logwtee_"+modelname+".txt")
    print("Open File Success")
except:
    print("Failed to Open File")
```

```python
i_avg_loss = []
i_mAP = []
```

```python
hasil_rl = file.readline()
while True:
    if "avg loss" in hasil_rl:
        # get indeks
        indeks = hasil_rl[1:hasil_rl.rfind(":")]
        # get avg loss
        raw_avg_loss = hasil_rl[hasil_rl.rfind(":")+1:hasil_rl.rfind(" avg loss")]
        avg_loss = raw_avg_loss[raw_avg_loss.rfind(",")+2:raw_avg_loss.rfind(" avg loss")]
        # save indeks dan avg loss
        i_avg_loss.append([indeks,avg_loss])
    if "mean average precision (mAP@0.50)" in hasil_rl:
        # get mAP
        mAP = hasil_rl[hasil_rl.rfind("=")+2:hasil_rl.rfind(",")]
        # save indeks dan mAP
        i_mAP.append([indeks,mAP])
    hasil_rl = file.readline()
    if "If you want to train from the beginning, then use flag in the end of training command: -clear " in hasil_rl:
        break
```

```python
# Import library for Graph
import matplotlib.pyplot as plt
```

```python
# line 1 average loss 
x1 = []
y1 = []
for i in i_avg_loss:
    x = i[0]
    y = i[1]#[:i[1].rfind(',')]
    x1.append(int(x))
    y1.append(float(y))
# plotting the average loss
plt.plot(x1, y1, lw = 0.5, label = "average loss")
 
# naming the x axis
plt.xlabel('batch')
# naming the y axis
plt.ylabel('value')
# giving a title to my graph
plt.title('Average Loss')

# show a legend on the plot
plt.legend()

figAvgErr = plt.gcf()

# save plot
figAvgErr.savefig(modelname+"-AvgErr.png", dpi = 300)

# set plot size
#plt.figure(figsize = (10,10))

# function to show the plot
plt.show()
```

```python
# line 2 mAP
x2 = []
y2 = []
for i in i_mAP:
    x = i[0]
    y = i[1]
    x2.append(int(x))
    y2.append(float(y))
# plotting the mAP
plt.plot(x2, y2, lw = 0.5, label = "mAP")

# naming the x axis
plt.xlabel('batch')
# naming the y axis
plt.ylabel('value')
# giving a title to my graph
plt.title('Mean Average Precision')
 
# show a legend on the plot
plt.legend()

figmAP = plt.gcf()
# save plot
figmAP.savefig(modelname+"-mAP.png", dpi = 300)

# function to show the plot
plt.show()
```
